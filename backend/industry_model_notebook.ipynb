```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("IndustryModelDataset.csv")
df.head()
```

```python
df.info()
```

```python
df = df.loc[:, ["NumberofFloors", "PropertyGFAParking", "PropertyGFABuilding(s)",
          "SiteEUI(kBtu/sf)", "SourceEUI(kBtu/sf)", "SteamUse(kBtu)", "Electricity(kBtu)",
          "NaturalGas(kBtu)", "TotalGHGEmissions"]]

df.dropna(inplace=True)

df.info()
```

```python
x_feat = [i for i in df.columns if i!= "TotalGHGEmissions"]
y_feat = "TotalGHGEmissions"

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.loc[:, x_feat], df[y_feat], test_size = 0.2)

x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(max_depth=100, random_state=0)

clf.fit(x_train, y_train)

print(f"Accuracy on test data: {clf.score(x_test, y_test)*100}%")
```

```python
import pickle

filename = "industry_RF_model.pickle"

# save model
pickle.dump(clf, open(filename, "wb"))
```
